## **Lowest Risk Majors**

```python
spread_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
```

```python
clean_df.insert(1, 'Spread', spread_col)
```

```python
clean_df.head()
```

```python
low_risk = clean_df.sort_values('Spread')
```

```python
low_risk[['Undergraduate Major', 'Spread']].head()
```

## **Majors with the Greatest Spread in Salaries**

```python
high_spread = clean_df.sort_values('Spread', ascending=False)
```

```python
high_spread[['Undergraduate Major', 'Spread']].head()
```

```python
highest_spread = clean_df.sort_values('Mid-Career Median Salary', ascending=False)
```

```python
highest_spread[['Undergraduate Major', 'Mid-Career Median Salary']].head()
```

## **Highest Potential Majors**

```python
high_potential = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
```

```python
high_potential[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

## **Grouping in STEM, Business or HASS (Humanities, Arts, and Social Science)**

```python
clean_df.groupby('Group').count()
```

## **Average Salary by Group**

```python
clean_df.groupby('Group').mean()
```
